## Checking for outliers

```python
from scipy.stats import zscore

# Assuming df_train is your DataFrame and numerical_features is a list of your numerical columns
z_scores = df_train[numerical_features].apply(zscore)

# Identify outliers where the z-score is > 3 or < -3
outliers = (z_scores > 3) | (z_scores < -3)

# Print out the number of outliers per feature
print("Outliers per feature:")
print(outliers.sum())
```

```python
# List of continuous features that might need capping
continuous_features = [
    'age', 'alcohol_consumption_per_week', 'physical_activity_minutes_per_week',
    'diet_score', 'sleep_hours_per_day', 'screen_time_hours_per_day', 
    'bmi', 'waist_to_hip_ratio', 'systolic_bp', 'diastolic_bp', 'heart_rate', 
    'cholesterol_total', 'hdl_cholesterol', 'ldl_cholesterol', 'triglycerides'
]

# Loop through each feature to cap outliers
for feature in continuous_features:
    lower_percentile = df_train[feature].quantile(0.01)  # 1st percentile
    upper_percentile = df_train[feature].quantile(0.99)  # 99th percentile

    # Cap outliers below the 1st percentile and above the 99th percentile
    df_train[feature] = np.where(df_train[feature] < lower_percentile, lower_percentile, df_train[feature])
    df_train[feature] = np.where(df_train[feature] > upper_percentile, upper_percentile, df_train[feature])
```

```python
# List of features we capped
features_to_visualize = ['bmi', 'physical_activity_minutes_per_week', 'cholesterol_total', 'triglycerides', 'systolic_bp']

# Plotting box plots for these features
plt.figure(figsize=(15, 10))

for i, feature in enumerate(features_to_visualize, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=df_train[feature])
    plt.title(f'Box Plot for {feature}')
    plt.tight_layout()

plt.show()
```

# Feature Engineering

```python
processed_train = df_train
processed_test = df_test
```

```python
def age_group(age):
    if age < 30:
        return 'Young'
    elif age < 50:
        return 'Middle-Aged'
    else:
        return 'Old'
        
processed_train['age_group'] = processed_train['age'].apply(age_group)
processed_test['age_group'] = processed_test['age'].apply(age_group)
```

```python
def bmi_category(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif bmi < 24.9:
        return 'Normal'
    elif bmi < 29.9:
        return 'Overweight'
    else:
        return 'Obese'
        
processed_train['bmi_category'] = processed_train['bmi'].apply(bmi_category)
processed_test['bmi_category'] = processed_test['bmi'].apply(bmi_category)
```

```python
# Function to categorize blood pressure
def categorize_blood_pressure(row):
    if row['systolic_bp'] < 120 and row['diastolic_bp'] < 80:
        return 'Normal'
    elif (120 <= row['systolic_bp'] < 140) or (80 <= row['diastolic_bp'] < 90):
        return 'Pre-hypertension'
    else:
        return 'Hypertension'

# Apply the function to create a new column for blood pressure category
processed_train['blood_pressure_category'] = processed_train.apply(categorize_blood_pressure, axis=1)
processed_test['blood_pressure_category'] = processed_test.apply(categorize_blood_pressure, axis=1)
```

```python
# Create LDL/HDL ratio
processed_train['ldl_hdl_ratio'] = processed_train['ldl_cholesterol'] / processed_train['hdl_cholesterol']
processed_test['ldl_hdl_ratio'] = processed_test['ldl_cholesterol'] / processed_test['hdl_cholesterol']
```

```python
def categorize_cholesterol(row):
    if row['cholesterol_total'] < 200:
        return 'Normal'
    elif 200 <= row['cholesterol_total'] < 240:
        return 'Borderline High'
    else:
        return 'High'

processed_train['cholesterol_risk'] = processed_train.apply(categorize_cholesterol, axis=1)
processed_test['cholesterol_risk'] = processed_test.apply(categorize_cholesterol, axis=1)
```

```python
def categorize_activity(minutes):
    if minutes < 150:
        return 'Low Activity'
    elif 150 <= minutes <= 300:
        return 'Moderate Activity'
    else:
        return 'High Activity'

processed_train['activity_category'] = processed_train['physical_activity_minutes_per_week'].apply(categorize_activity)
processed_test['activity_category'] = processed_test['physical_activity_minutes_per_week'].apply(categorize_activity)
```

```python
def categorize_diet(score):
    if score < 3:
        return 'Poor Diet'
    elif 3 <= score <= 7:
        return 'Average Diet'
    else:
        return 'Healthy Diet'

processed_train['diet_category'] = processed_train['diet_score'].apply(categorize_diet)
processed_test['diet_category'] = processed_test['diet_score'].apply(categorize_diet)
```

```python
def categorize_sleep(hours):
    if hours < 6:
        return 'Short Sleep'
    elif 6 <= hours <= 8:
        return 'Adequate Sleep'
    else:
        return 'Long Sleep'

processed_train['sleep_category'] = processed_train['sleep_hours_per_day'].apply(categorize_sleep)
processed_test['sleep_category'] = processed_test['sleep_hours_per_day'].apply(categorize_sleep)
```

```python
def categorize_screen_time(hours):
    if hours < 2:
        return 'Low Screen Time'
    elif 2 <= hours <= 4:
        return 'Moderate Screen Time'
    else:
        return 'High Screen Time'

processed_train['screen_time_category'] = processed_train['screen_time_hours_per_day'].apply(categorize_screen_time)
processed_test['screen_time_category'] = processed_test['screen_time_hours_per_day'].apply(categorize_screen_time)
```

```python
# Create the 'lifestyle_risk' column
def categorize_lifestyle(row):
    # Low Risk: High Activity, Healthy Diet, Normal BMI, Non-smoker
    if row['activity_category'] == 'High Activity' and row['diet_category'] == 'Healthy Diet' and row['bmi_category'] == 'Normal' and row['smoking_status'] != 'Smoker':
        return 'Low Risk'
    # Medium Risk: Moderate Activity or Average Diet or Overweight BMI or Former smoker
    elif (row['activity_category'] == 'Moderate Activity' or row['diet_category'] == 'Average Diet' or row['bmi_category'] == 'Overweight') or row['smoking_status'] == 'Former smoker':
        return 'Medium Risk'
    # High Risk: Low Activity, Poor Diet, Obese BMI, or Smoker
    else:
        return 'High Risk'

# Apply the function to the DataFrame
processed_train['lifestyle_risk'] = processed_train.apply(categorize_lifestyle, axis=1)
processed_test['lifestyle_risk'] = processed_test.apply(categorize_lifestyle, axis=1)
```

```python
def categorize_medical_risk(row):
    # Count the number of medical conditions present
    conditions = [row['family_history_diabetes'], row['hypertension_history'], row['cardiovascular_history']]
    risk_count = sum(conditions)  # Count how many conditions are present
    if risk_count == 0:
        return 'Low Risk'  # No conditions
    elif risk_count == 1:
        return 'Medium Risk'  # One condition
    else:
        return 'High Risk'  # Two or more conditions

# Apply the function to the DataFrame
processed_train['medical_risk'] = processed_train.apply(categorize_medical_risk, axis=1)
# Apply the function to the testing dataset
processed_test['medical_risk'] = processed_test.apply(categorize_medical_risk, axis=1)
```

```python
def overall_risk(row):
    if row['lifestyle_risk'] == 'Low Risk' and row['medical_risk'] == 'Low Risk':
        return 'Low Risk'
    elif row['lifestyle_risk'] == 'High Risk' or row['medical_risk'] == 'High Risk':
        return 'High Risk'
    else:
        return 'Medium Risk'

processed_train['overall_risk'] = processed_train.apply(overall_risk, axis=1)
processed_test['overall_risk'] = processed_test.apply(overall_risk, axis=1)
```

### Scaling the Data

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Scale the numerical features in the training set
X_scaled = X[numerical_cols].copy()
X_scaled[numerical_cols] = scaler.fit_transform(X_scaled)

# Scale the numerical features in the test set (use the same scaler as the training set)
X_test_scaled = X_test[numerical_cols].copy()
X_test_scaled[numerical_cols] = scaler.transform(X_test_scaled)

# Combine the scaled numerical columns with the original categorical columns
X_train_final = X_scaled.join(X[categorical_cols])
X_test_final = X_test_scaled.join(X_test[categorical_cols])

# Check final shapes after scaling
print(f"X_train_final shape: {X_train_final.shape}")
print(f"X_test_final shape: {X_test_final.shape}")
```

```python
from sklearn.model_selection import train_test_split

# Train-validation split
X_train, X_val, y_train, y_val = train_test_split(
    X_train_final, y, test_size=0.2, random_state=42, stratify=y
)

print(f"\nTRAIN-VALIDATION SPLIT:")
print(f"  Training set: {X_train.shape[0]:,} samples ({len(X_train)/len(X)*100:.0f}%)")
print(f"  Validation set: {X_val.shape[0]:,} samples ({len(X_val)/len(X)*100:.0f}%)")
```

# Training the model 

## Using Logistic Regression

```python
# Imports for Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
```

```python
log_reg = LogisticRegression(
    penalty="l2",          
    solver="liblinear",
    max_iter=300,
    random_state=42
)

param_distributions = {
    "C": np.logspace(-2, 1, 10),   
    "class_weight": [None, "balanced"]
}

random_search = RandomizedSearchCV(
    estimator=log_reg,
    param_distributions=param_distributions,
    n_iter=5,              
    scoring="roc_auc",
    cv=2,                  
    n_jobs=-1,
    random_state=42,
    verbose=1
)
```

```python
random_search.fit(X_train, y_train)
best_model = random_search.best_estimator_
print("Best params:", random_search.best_params_)
```

```python
y_val_pred = best_model.predict(X_val)
y_val_proba = best_model.predict_proba(X_val)[:, 1]

print("VALIDATION METRICS (Logistic Regression)")
print("--------------------------------------")
print(f"ROC-AUC: {roc_auc_score(y_val, y_val_proba):.4f}\n")

print("Classification Report:")
print(classification_report(y_val, y_val_pred))
```

```python
probs = best_model.predict_proba(X_val)[:, 1]

for t in [0.4, 0.35, 0.3, 0.25]:
    preds = (probs >= t).astype(int)
    print(f"\nThreshold = {t}")
    print(classification_report(y_val, preds))
```

In this section, a Logistic Regression model was trained and tuned using RandomizedSearchCV, resulting in the best parameters: C ≈ 0.215 and class_weight=None. The model achieved a ROC-AUC of 0.696 on the validation set. A decision threshold of 0.4 was applied to increase the model’s sensitivity to diabetes. With this threshold, the model correctly identified 96% of diabetic patients, while many healthy patients were also flagged, which is acceptable for a screening scenario. Feature importance analysis highlighted the top predictors, providing insight into which variables contributed most to the model’s predictions. Overall, the model effectively serves as a screening tool, prioritizing the detection of diabetes cases.
